```python
X = card.drop("Class", axis =1)
Y = card["Class"]
```

MACHINE LEARNING MODEL

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2, random_state =2)
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
ytrainpredict = model.predict(X_train)
train_data = accuracy_score(ytrainpredict,Y_train)
print(train_data)
```

```python
ytestpredict = model.predict(X_test)
test_data = accuracy_score(ytestpredict,Y_test)
print(test_data)
```

```python
train_matrix = confusion_matrix(Y_train, ytrainpredict)
test_matrix = confusion_matrix(Y_test, ytestpredict)
print("train_matrix",train_matrix)
print("test_matrix:",test_matrix)
```

```python
train_report = classification_report(Y_train, ytrainpredict)
test_report = classification_report(Y_test, ytestpredict)
print("Train Report:",train_report)
print("Test Report:",test_report)
```
